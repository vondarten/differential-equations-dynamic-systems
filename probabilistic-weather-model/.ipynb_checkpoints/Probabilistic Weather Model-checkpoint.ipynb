```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
A = np.array([[0.5, 0.5, 0.25], [0.25, 0.0, 0.25], [.25, .5, .5]])
```

```python
print(A)
```

```python
x_today = np.array([[1], [0], [0]])
```

```python
weather_over_time = np.zeros((3, 50))

for i in range(50):
    x_tomorrow = A@x_today
    weather_over_time[:, i] = x_tomorrow
    x_today = x_tomorrow
```
