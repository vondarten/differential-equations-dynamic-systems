Taylor Series are widely used to approximate continuous functions. 
It can be obtained by the following expression:

$$f(x)=\sum_{k=0}^\infty f^{(k)}(a)\frac{(x-a)^k}{k!}$$

Here follows a python code that is able to visualize Taylor's expansion by varying the number of components summed up.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
from scipy.special import factorial
```

```python
# interval: x
x = np.linspace(-20, 20, 2000)
y = np.sin(x)

plt.ylim(-20,20)
plt.plot(x, y, 'k', linewidth=2)

plt.grid(True)

# First order expansion

# p holds all coeficientes of the Polynomial

# x + 0
p = [1,0] 
y_first = np.polyval(p, x)
plt.plot(x, y_first, 'y--', linewidth=1.2)

# Third order expansion

p = [-1, factorial(3), 0, 1, 0] 
y_third = np.polyval(p, x)
plt.plot(x, y_third, 'g--', linewidth=1.2)

# Fifth order expansion

p=[1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
y_fifth = np.polyval(p,x)
plt.plot(x, y_fifth, 'b--', linewidth=1.2)

# Seventh order expansion
p = [-1/factorial(7), 0, 1/factorial(5), 0, -1/factorial(3), 0, 1,0]
y_seventh = np.polyval(p,x)
plt.plot(x, y_seventh, 'r--', linewidth=1.2)
```

When higher-level terms are added up, the approximation tends clearly to the original function: sin(x)
